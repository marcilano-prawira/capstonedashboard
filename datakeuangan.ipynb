#FEATURE ENGINEERING

Pembobotan score status kesiapan :
1. Score Dana darurat (MAX) : 40 poin
2. Score Rasio Utang (MAX) : 30 poin
3. Score Cash Flow (MAX) : 30 poin
- Max Score : 100 poin

```python
df_keuangan = df_clean_data.copy()
```

##Score Dana darurat

```python
df_keuangan['Rasio Dana Darurat'] = (df_keuangan['emergency_fund'] / df_keuangan['monthly_expense_total']).round(2)

# jika rasio >= 6, dapat 40 poin, jika <6 poin dihitung proporsional.
df_keuangan['score_dana_darurat'] = np.where(
    df_keuangan['Rasio Dana Darurat'] >= 6,
    40,
    (df_keuangan['Rasio Dana Darurat'] / 6) * 40
).round(2)
```

##Score Beban Utang

```python
# semakin kecil utang, semakin baik. jika utang >= 80% dari gaji, skornya 0.
df_keuangan['score_utang'] = np.where(
    df_keuangan['debt_to_income_ratio'] >= 0.8,
    0,
    30 - (df_keuangan['debt_to_income_ratio'] / 0.8 * 30)
).round(2)
```

##Score Cash Flow

```python
# pemasukan dikurangi Pengeluaran. dan uang sisa harus positif
df_keuangan['cash_flow'] = df_keuangan['monthly_income'] - df_keuangan['monthly_expense_total']

# jika cash flow lebih dari 0, dapat 30 poin. jika minus/pas-pasan, dapat 0 poin.
df_keuangan['score_cashflow'] = np.where(df_keuangan['cash_flow'] > 0, 30, 0).round(2)
```

##Total Score

```python
df_keuangan['financial_health_score'] = (
    df_keuangan['score_dana_darurat'] +
    df_keuangan['score_utang'] +
    df_keuangan['score_cashflow']
).round(2)
```

#Status Kesiapan Investasi

Paramater Treshold (ambang batas skor)

```python
def tentukan_status(skor):
    if skor >= 60:
        return 'Siap Investasi'
    elif skor >= 40:
        return 'Perbaiki Dana Darurat dan pelunasan Utang'
    else:
        return 'WARING!!! : Fokus Lunasi Utang & Nabung'

df_keuangan['status_kesiapan'] = df_keuangan['financial_health_score'].apply(tentukan_status)
```
